```python
class Person:
  def __init__(self,fname,lname):
    self.fname = fname
    self.lname = lname

  def welcome(self):
    return 'welcome '+self.fname+" "+self.lname

class Student(Person):
  def __init__(self,fname,lname,age):
    super().__init__(fname,lname)
    self.age = age

  def swel(self):
    return " this object is a student"
  

p1 = Person('john','francis')
s1 = Student('ben',"10",'3')

print(p1.swel())
```
