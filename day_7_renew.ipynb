```python
import random 


lower = 'qwertyuiopasdfghhjklzxcvbnm'
upper = 'QWERTYUIOPASDFGHJKLZXCVBNM'
sym = '!@#$%^&&(){}:<>/.,;[]'
num = '1234567890'

all = lower + upper + sym + num

length = 34

password = ''.join(random.sample(all,length))

print(password)
```

```python
#class 
class Employee:
   #instance
  def __init__(self,fname,lname):
    self.first_name = fname
    self.last_name = lname
#methods
  def fullname(self):
    return self.first_name + " " +self.last_name

  def welcom(self):
    return "welcome mr. " +self.first_name + " " +self.last_name

#objects
emp1 = Employee('john','francis')

emp2 = Employee('gunal','dheena')



print(emp2.welcom())
```
