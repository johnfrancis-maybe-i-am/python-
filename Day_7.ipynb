```python
import random

upper = "QWETYUIOPASFDFGJKLZXBNMM"
lower = "qwertyuiopasdfghjklzxcbbnm"
num = "0123456789"
sc = "!@#$%^&*()<>?:"

all = upper+lower+num+sc
length = input("Enter the password lenght: ")
length = int(length)

password = "".join(random.sample(all,length))

print(password)
```

```python
class Employee: #class

  def __init__(self,fname,lname,email): #instance
    self.fname = fname
    self.lname = lname
    self.email = email

  def fullname(self): #method
    return self.fname +" "+self.lname


employee1 = Employee("john","francis","john@gmail.com")  #objects
employee2 = Employee("ponar","","ponar@gmail.com")
employee3 = Employee("ben","tenison","ben@gmail.com")

print(employee3.fullname())



  
  
```
